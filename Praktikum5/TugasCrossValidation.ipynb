```python
import pandas as pd
from sklearn.model_selection import cross_validate, KFold
from sklearn.ensemble import RandomForestClassifier
```

```python
train_url = 'https://raw.githubusercontent.com/RIZ4N/RizanOesvanIdrati_2411531010_ML2526/refs/heads/main/Praktikum5/train.csv'
test_url = 'https://raw.githubusercontent.com/RIZ4N/RizanOesvanIdrati_2411531010_ML2526/refs/heads/main/Praktikum5/test.csv'

train_df = pd.read_csv(train_url)
test_df  = pd.read_csv(test_url)
```

```python
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
train_df['Embarked'] = train_df['Embarked'].fillna(train_df['Embarked'].mode()[0])
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].median())
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].median())
```

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
X_train = pd.get_dummies(train_df[features], drop_first=True)
y_train = train_df['Survived']
X_test = pd.get_dummies(test_df[features], drop_first=True)
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

```python
model = RandomForestClassifier(random_state=42)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scoring_metrics = ['accuracy', 'precision', 'recall', 'f1']
cv_results = cross_validate(model, X_train, y_train, cv=kf, scoring=scoring_metrics)
```

```python
print("=== Analisis Kinerja Model (5-Fold Cross Validation) ===")
print(f"Rata-rata Akurasi  : {cv_results['test_accuracy'].mean():.4f}")
print(f"Rata-rata Presisi  : {cv_results['test_precision'].mean():.4f}")
print(f"Rata-rata Recall   : {cv_results['test_recall'].mean():.4f}")
print(f"Rata-rata F1-Score : {cv_results['test_f1'].mean():.4f}\n")
```

```python
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
submission = pd.DataFrame({'PassengerId': test_df['PassengerId'], 'Survived': predictions})
print(submission.head())
submission.to_csv('submission.csv', index=False)
print("File submission.csv")
```
